## Data Ingestion -  DocumentLoader

https://python.langchain.com/docs/integrations/document_loaders/

### Text loader


```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
loader
```

```python
text_document = loader.load()
text_document
```

### PDF loader

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Speech.pdf")

pdf = loader.load()
pdf
```

### Webpages

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(web_paths=("https://www.w3schools.com/python/python_lists.asp",),)
web = loader.load()
web
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_paths=("https://www.w3schools.com/python/python_lists.asp",),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=("w3-example")
                       )))
web = loader.load()
web
```

### ArXiv

```python
from langchain_community.document_loaders import ArxivLoader

lodaer = ArxivLoader(query="1706.03762",load_max_docs=2)

arxiv_doc = lodaer.load()
arxiv_doc
```

### Wikipedia

```python
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query="Indian Prime Minister",load_max_docs=3)

wiki_doc = loader.load()
wiki_doc
```
