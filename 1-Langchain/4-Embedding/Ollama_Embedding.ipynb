## Ollama
##### Ollama supports embedding models, making it possible to build retrieval augmented generation(RAG) appllications that combine text prompts with exisiting documents or other data

- To donwload ollama, go to URL: https://ollama.com/download and donwload it
- Then install it and follow the default steps
- open CMD and run command : ollama run gemma:2b (wait for sometime till it pull gemma:2b model locally, llama2 model comes as default no need to donwload it)
- Just to test, type "who are you?" and check if it is responding that => I am a large language model, trained by Google. 

```python
## Default model is llama2

from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings()

embeddings
```

```python
## Try model: gemma:2b
embeddings = (OllamaEmbeddings(model="gemma:2b"))

embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "I am Indian student",
        "I am German student"
    ]
)
```

```python
r1[0]
```

```python
len(r1[0])
```

#### Try different Models
- You can look for different models here at : https://ollama.com/blog/embedding-models
- Run this in CMD to pull model=> ollama pull mxbai-embed-large

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document"
query_result = embeddings.embed_query(text)
query_result
```
